## 1.Регулярные Выражения
Если вкратце, регулярные выражения предаставляют из себя шаблоны, которые "накладывают" на текст, чтобы найти искомые слова, номера и т.п
Примеры использования регулярных выражений вы найдёте по [ссылке](https://habr.com/ru/post/349860/) 
Разберём на примере их применение на практике
```
import re # импортируем библиотеку, она встроенная в python
myString = '+38099453456Alexander,+3806731537454Bohdan'
# допустим, мы хотим получить список из номеров телефонов, чтобы наш робот
# отправил туда свою рассылку
re.findall('\+[0-9]+',myString) 
# 1 аругмент - регулярное выражение, второй - строка
# OUT: ['+38099453456', '+3806731537454']
``` 
Если вам нужно получить только один элемент из поиска, обратитесь к нему по индексу:
```
re.findall('\+[0-9]+',myString)[0]
# OUT: '+38099453456'
```
## 2. Математика
##Форумула косинусного расстояния:
cosine.png
## Затруднения?
В числителе сложите все элементы ПОЧЛЕННОГО перемножения векторов. Если они получились разной размерности - вы допустили ошибку!
В знаменателе перемножьте корни из суммы квадратов всех элементов векторов. Другими словами: перемножьте нормы векторов
